**FILTROS BASICOS CON CONECTORES LOGICOS**

```python
#Filtro Y
filtro7=df[(df["independence_year"] > 1900) & (df["region"] == "South America")]
filtro7
```

Para el filtro 7 se muestran los países cuya independencia ocurrió después de 1900 y en la región de Sudamérica

```python
#Filtro o
filtro8=df[(df["capital"] > 2300) | (df["gnp"] > 800)]
filtro8
```

Para el filtro 8 se muestran los países cuya capital sea mayor a 2300 o que su gnp sea mayor a 800

```python
#Convertir archivo filtrado a csv
filtro1.to_csv("Actividad1_Filtro8.csv")
```

```python
#Descargar archivo filtrado en csv
from google.colab import files
files.download("Actividad1_Filtro8.csv")
```

```python
#Filtro not
filtro9=df[(df["population"] > 100000) & ~(df["life_expectancy"] < 70)]
filtro9
```

Para el filtro 9 se muestran los países cuya población sea mayor a 100000 pero que su expectativa de vida no sea menor a 70

```python
#Convertir archivo filtrado a csv
filtro9.to_csv("Actividad1_Filtro9.csv")
```
