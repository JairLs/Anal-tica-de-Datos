```python
df= pd.read_csv("CountryTable.csv", index_col=2)
df
```

```python
#Filtro filas vs columnas
filtro4=df.loc[["Asia","Europe"],["life_expectancy"]]
filtro4
```

```python
#Filtro por cabecera
filtro5=df.head(20)
filtro5
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparacion
filtro6=df[df["surface_area"] > 500000]
filtro6
```

**FILTROS BASICOS CON CONECTORES LOGICOS**

```python
#Filtro Y
filtro7=df[(df["independence_year"] > 1900) & (df["region"] == "South America")]
filtro7
```

```python
#Filtro o
filtro8=df[(df["capital"] > 2300) | (df["gnp"] > 800)]
filtro8
```

```python
#Filtro not
filtro9=df[(df["population"] > 100000) & ~(df["life_expectancy"] < 70)]
filtro9
```

```python
#Convertir archivo filtrado a csv
filtro9.to_csv("Actividad1.csv")
```
